```python
def subplt(img1, img2, img3, img4, img5):
  fig = plt.figure()
  ax = fig.add_subplot(2, 3, 1)
  imgplot = plt.imshow(img1)
  plt.axis('off')
  ax = fig.add_subplot(2, 3, 2)
  imgplot = plt.imshow(img2)
  plt.axis('off')
  ax = fig.add_subplot(2, 3, 3)
  imgplot = plt.imshow(img3)
  plt.axis('off')
  ax = fig.add_subplot(2, 3, 4)
  imgplot = plt.imshow(img4)
  plt.axis('off')
  ax = fig.add_subplot(2, 3, 5)
  imgplot = plt.imshow(img5)
  plt.axis('off')
```

### Applying pseudocolor schemes to image plots

```python
def subplt1(img1, img2, img3, img4, img5):
  fig = plt.figure()
  ax = fig.add_subplot(2, 3, 1)
  imgplot = plt.imshow(img1)
  plt.axis('off')
  ax = fig.add_subplot(2, 3, 2)
  imgplot = plt.imshow(img2)
  plt.axis('off')
  ax = fig.add_subplot(2, 3, 3)
  imgplot = plt.imshow(img3)
  plt.axis('off')
  ax = fig.add_subplot(2, 3, 4)
  imgplot = plt.imshow(img4)
  plt.axis('off')
  ax = fig.add_subplot(2, 3, 5)
  imgplot = plt.imshow(img5)
  plt.axis('off')
```

```python
lum_img1 = img1[:, :, 0]
lum_img2 = img2[:, :, 0]
lum_img3 = img3[:, :, 0]
lum_img4 = img4[:, :, 0]
lum_img5 = img5[:, :, 0]
subplt1(lum_img1, lum_img2, lum_img3, lum_img4, lum_img5)
```

```python
def subplt2(img1, img2, img3, img4, img5, cmp):
  fig = plt.figure()
  ax = fig.add_subplot(2, 3, 1)
  imgplot = plt.imshow(img1, cmap = cmp)
  plt.axis('off')
  ax = fig.add_subplot(2, 3, 2)
  imgplot = plt.imshow(img2, cmap = cmp)
  plt.axis('off')
  ax = fig.add_subplot(2, 3, 3)
  imgplot = plt.imshow(img3, cmap = cmp)
  plt.axis('off')
  ax = fig.add_subplot(2, 3, 4)
  imgplot = plt.imshow(img4, cmap = cmp)
  plt.axis('off')
  ax = fig.add_subplot(2, 3, 5)
  imgplot = plt.imshow(img5, cmap = cmp)
  plt.axis('off')
```

### With a luminosity (2D, no color) image
### color map = "hot"

```python
subplt2(lum_img1, lum_img2, lum_img3, lum_img4, lum_img5, "hot")
```

### color map = "nipy_spectral"

```python
subplt2(lum_img1, lum_img2, lum_img3, lum_img4, lum_img5, "nipy_spectral")
```

### Color scale reference

```python
def subplt3(img1, img2, img3, img4, img5):
  fig = plt.figure()
  ax = fig.add_subplot(2, 3, 1)
  imgplot = plt.imshow(img1)
  plt.colorbar()
  plt.axis('off')
  ax = fig.add_subplot(2, 3, 2)
  imgplot = plt.imshow(img2)
  plt.colorbar()
  plt.axis('off')
  ax = fig.add_subplot(2, 3, 3)
  imgplot = plt.imshow(img3)
  plt.colorbar()
  plt.axis('off')
  ax = fig.add_subplot(2, 3, 4)
  imgplot = plt.imshow(img4)
  plt.colorbar()
  plt.axis('off')
  ax = fig.add_subplot(2, 3, 5)
  imgplot = plt.imshow(img5)
  plt.colorbar()
  plt.axis('off')
```

```python
subplt3(lum_img1, lum_img2, lum_img3, lum_img4, lum_img5)
```

### Converts the data to a single channel containing floating-point grayscale color

```python
pip install toyplot
```

```python
import PIL.Image
import toyplot
import skimage.color

gray1 = skimage.color.rgb2gray(img1)
gray1.shape, gray1.dtype
```

```python
gray2 = skimage.color.rgb2gray(img2)
gray2.shape, gray2.dtype
```

```python
gray3 = skimage.color.rgb2gray(img3)
gray3.shape, gray3.dtype
```

```python
gray4 = skimage.color.rgb2gray(img4)
gray4.shape, gray4.dtype
```

```python
gray5 = skimage.color.rgb2gray(img5)
gray5.shape, gray5.dtype
```

```python
canvas, mark = toyplot.image(gray1, width=300)
canvas, mark = toyplot.image(gray2, width=300)
canvas, mark = toyplot.image(gray3, width=300)
canvas, mark = toyplot.image(gray4, width=300)
canvas, mark = toyplot.image(gray5, width=300)
```

### Displaying one-bit image

```python
import skimage.filters
binary1 = gray1 > skimage.filters.threshold_isodata(gray1)
binary1.shape, binary1.dtype
canvas, mark = toyplot.image(binary1, width=300)
```

```python
# Grayscale data can also be mapped using Toyplot’s usual Color Mapping functionality
colormap = toyplot.color.linear.map("Blackbody")
canvas, mark = toyplot.image((gray1, colormap), width=300)
canvas, mark = toyplot.image((gray2, colormap), width=300)
canvas, mark = toyplot.image((gray3, colormap), width=300)
canvas, mark = toyplot.image((gray4, colormap), width=300)
canvas, mark = toyplot.image((gray5, colormap), width=300)
```
